After working with Random Forests and MLPs, the next step is to try a more complex ML algorithm. I chose LSTM because I have experience working with it, and it seemed like one of the most common deep learning networks for many-to-one sequential regression.

## LSTM

One advantage of the LSTM over previous models is the input data it takes. In our many-to-one regression problem, for a given observation window we have an entire matrix of sequential feature data, and only two output scalar values to predict. For a simple model like RF, this requires condensing that feature set for each window by flattening the data and then using PCA or some other method to aggregate it. LSTM, on the other hand, can take a 3D tensor input. So our X matrix is of shape (num_samples, sequence_length, num_features). In this case, that looks like (11536, 30, 13). We feed that input data to the LSTM, and then it outputs a target of (num_samples, 2) for our two target variables.


I developed an LSTM with the following hyperparameters and began testing:
- Hidden size: The number of neurons in each hidden layer
- Number of layers: The number of hidden layers (layers between the input and output)
- Dropout: The ratio of neurons to temporarily set to zero during a training run
- Weight decay: Similar to L2 regularization, this adds a penalty term to the loss function to force model weights to remain small

The first two layers determine the complexity, while the second two determine the level of regularization. Be striking a balance between these two factors, we hope to create a successful model.

The pipeline trains for 500 epochs with a patience of 40 for early stopping.

### Results

Unfortunately, no combination of hyperparameters produced a validation RMSE of less than 6.5. I didn't run a grid search on the model as it took a while to train, but in all the experiments I did, the RMSE hovered in a very tight range around 6.5 to 6.7.
