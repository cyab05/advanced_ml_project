## MLP with PCA

From the last models, PCA didn't seem compatible with Random Forest, at least at the low-rank approximation we were using. So we turn to a simple multilayer perceptron or MLP. This is the term for the most simple neural networks, where each neuron has a simple activation function and nothing more. The thought process with combining PCA and an MLP was that the PCA would eliminate the noise from our data and the MLP, while 'simple' when compared to deep-learning, would have more complexity than the random forest and be able to parse the signal from the data.

This being the first neural network we have trained, we set up a training loop with proper hyperparameter and validation tracking. For ease of use, we connect the code to MLFlow, which allows us to easily track the hyperparameters of the best models. It even helps us make pretty charts easily, like this one!

### Initial Grid Search

We implemented a grid search using MLFlow, and tested a range of k-values for the PCA, number of hidden layers, hidden layer sizes, weight decay values ($L_2$ regularization), and dropouts. My initial values ranges for each of these was pretty standard: 8 to 64 for the hidden layer sizes, 0 to 0.4 for the dropouts, etc. For the PCA, remembering the Scree Plot with an elbow around 6 or 7, I choose small values of 10, 20, to 30, assuming these to be even too big. However, the grid search quickly revealed that the more complex models were best. The initial model I test was one with 4 hidden layers, 64 neurons each, 0.0 dropout, and a PCA value of 50.
